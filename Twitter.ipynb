<h3>One day data</h3>

```r
df5Apr$created_at <- as.Date(df5Apr$created_at)
TweetsOneDay <- df5Apr[df5Apr$created_at == '2018-04-04',]
TweetsOneDay<-TweetsOneDay[TweetsOneDay$is_retweet!="TRUE", ]

for(k in 1:as.integer(dim(TweetsOneDay)[1]/25)){    
    file=paste0("./data/TweetData/OneDay/tweet_",k,".","txt")    
    for(i in 25*(k-1)+1:25*(k)){
            write.table(TweetsOneDay[i,'text'], file = file,append = TRUE,row.names = FALSE, col.names=FALSE,quote=FALSE,sep = "\t",eol="\n")    
    }
}
```
